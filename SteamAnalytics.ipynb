```python
#cleaning up the data
def DataFrameScrubber(dfIn):
    
    for i in range(len(dfIn)):

        # Convert the release date to the number of days since release    
        if  type(dfIn.iloc[i,2]) == str:
            dfIn.iloc[i,2] = (datetime.datetime.now() - datetime.datetime.strptime( dfIn.iloc[i,2], "%m/%d/%Y")).days

        # Change column name to 'Days Since Release'    
        dfIn = dfIn.rename(columns = {'Release date':'Day Since Release'})

        if  type(dfIn.iloc[i,2]) == float:
            dfIn.iloc[i,2] = 999999

        # Scrub the price of a string
        # change it to 0 if it is free
        if dfIn.iloc[i,3] != 'Free':
            dfIn.iloc[i,3] = re.sub(r'[\D,.]', '',dfIn.iloc[i,3])
        else:
             dfIn.iloc[i,3] = 0

        #Scrub the metascoreand just make it a number
        if dfIn.iloc[i,4] != 'N/A':
            dfIn.iloc[i,4] = re.sub(r'[\%]','',dfIn.iloc[i,4].split()[0])
        if dfIn.iloc[i,4] == 'N/A':
             dfIn.iloc[i,4] = 999999

        #Take the First value of the number of users
        #Make it a number and remove commas
        S = dfIn.iloc[i,5]

        R1 = re.sub(r'[\,]', '',S)

        dfIn.iloc[i,5] = re.sub(r'[\D]',' ',R1).split()[0]

        #Scrub the playtime. Remove puntuation and median Value
        dfIn.iloc[i,6] = re.sub(r'[:]','',dfIn.iloc[i,6].split()[0])
        
    dfIn.iloc[:,2] = dfIn.iloc[:,2].astype(int)
    dfIn.iloc[:,3] = dfIn.iloc[:,3].astype(int)
    dfIn.iloc[:,4] = dfIn.iloc[:,4].astype(int)
    dfIn.iloc[:,5] = dfIn.iloc[:,5].astype(int)
    return dfIn
  
```
