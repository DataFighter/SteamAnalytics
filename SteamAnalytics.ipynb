```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from copy import copy
import re
%matplotlib inline
```

```python
dfStrategy = pd.read_csv('Strategy.CSV').drop('#', 1)
dfSports = pd.read_csv('Sports.CSV').drop('#', 1)
dfSimulation = pd.read_csv('Simulation.CSV').drop('#', 1)
dfRPG = pd.read_csv('RPG.CSV').drop('#', 1)
dfMMO = pd.read_csv('MMO.CSV').drop('#', 1)
dfIndie = pd.read_csv('Indie.CSV').drop('#', 1)
dfFree = pd.read_csv('Free.CSV').drop('#', 1)
dfEarly_Access = pd.read_csv('Early Access.CSV').drop('#', 1)
dfAdventure = pd.read_csv('Adventure.CSV').drop('#', 1)
dfAction = pd.read_csv('Action.CSV').drop('#', 1)
```

```python
dfAll = dfStrategy
dfAll = dfAll.merge(dfSports, how = 'outer')
dfAll = dfAll.merge(dfSimulation, how = 'outer')
dfAll = dfAll.merge(dfRPG, how = 'outer')
dfAll = dfAll.merge(dfMMO, how = 'outer')
dfAll = dfAll.merge(dfIndie, how = 'outer')
dfAll = dfAll.merge(dfFree, how = 'outer')
dfAll = dfAll.merge(dfEarly_Access, how = 'outer')
dfAll = dfAll.merge(dfAdventure, how = 'outer')
dfAll = dfAll.merge(dfAction, how = 'outer')
```

```python
#cleaning up the data

def DataFrameScrubber(dfIn):
    
    for i in range(len(dfIn)):

        # Convert the release date to the number of days since release    
        if  type(dfIn.iloc[i,1]) == str:
            dfIn.iloc[i,1] = (datetime.datetime.now() - datetime.datetime.strptime( dfIn.iloc[i,1], "%m/%d/%Y")).days

        # Change column name to 'Days Since Release'    
        dfIn = dfIn.rename(columns = {'Release date':'Day Since Release'})

        if  type(dfIn.iloc[i,1]) == float:
            dfIn.iloc[i,1] = 999999

        # Scrub the price of a string
        # change it to 0 if it is free
        if dfIn.iloc[i,2] != 'Free':
            dfIn.iloc[i,2] = re.sub(r'[\D,.]', '',dfIn.iloc[i,2])
        else:
             dfIn.iloc[i,2] = 0

        #Scrub the metascoreand just make it a number
        if dfIn.iloc[i,3] != 'N/A':
            dfIn.iloc[i,3] = re.sub(r'[\%]','',dfIn.iloc[i,3].split()[0])
        if dfIn.iloc[i,3] == 'N/A':
             dfIn.iloc[i,3] = 999999

        #Take the First value of the number of users
        #Make it a number and remove commas
        S = dfIn.iloc[i,4]

        R1 = re.sub(r'[\,]', '',S)

        dfIn.iloc[i,4] = re.sub(r'[\D]',' ',R1).split()[0]

        #Scrub the playtime. Remove puntuation and median Value
        dfIn.iloc[i,5] = re.sub(r'[:]','',dfIn.iloc[i,5].split()[0])
        
    dfIn.iloc[:,1] = dfIn.iloc[:,1].astype(int)
    dfIn.iloc[:,2] = dfIn.iloc[:,2].astype(int)
    dfIn.iloc[:,3] = dfIn.iloc[:,3].astype(int)
    dfIn.iloc[:,4] = dfIn.iloc[:,4].astype(int)
    dfIn.iloc[:,5] = dfIn.iloc[:,5].astype(int)
    return dfIn
  
```

```python
dfAllScrubbed = DataFrameScrubber(dfAll)
```

```python
dfAllScrubbed.to_csv('All.csv')
dfAllScrubbed
```
